# SE-ResNet50 EAST Training

**Важные параметры:**

## OHEM Ratio
- **`ohem_ratio`** - доля (fraction) тяжелых примеров для обучения
- **Диапазон: (0.0, 1.0]**
- **По умолчанию: 0.5** (50% самых сложных примеров)

**Примеры:**
- `ohem_ratio=0.3` → выбирает 30% самых сложных примеров
- `ohem_ratio=0.5` → выбирает 50% самых сложных примеров  
- `ohem_ratio=0.9` → выбирает 90% самых сложных примеров
- `ohem_ratio=1.5` → ❌ **ОШИБКА!** Нельзя выбрать 150% примеров!

**Рекомендации:**
- Для легких датасетов: `ohem_ratio=0.3-0.5`
- Для сложных датасетов: `ohem_ratio=0.7-0.9`
- Не используйте значения > 1.0!

```python
from manuscript.detectors import EAST


train_images = [r"C:\shared\data0205\data02065\ICDAR2015\train_images"] 
train_anns = [r"C:\shared\data0205\data02065\ICDAR2015\train.json"]
val_images = [r"C:\shared\data0205\data02065\ICDAR2015\test_images"]
val_anns = [r"C:\shared\data0205\data02065\ICDAR2015\test.json"]


# Запуск обучения
best_model = EAST.train(
    train_images=train_images,
    train_anns=train_anns,
    val_images=val_images,
    val_anns=val_anns,
    backbone_name="resnet101",
    model_name="east_resnet101",
    pretrained_backbone=True,
    ohem_ratio=0.9,
    target_size=1024,
    epochs=600,
    batch_size=1,
    lr=0.001,
    accumulation_steps=16,
    val_interval=1,
)
print("Сохранён чекпоинт:", best_model)
```
